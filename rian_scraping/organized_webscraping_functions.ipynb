## Function to Export as CSV

```python
import csv
import os


def data_to_csv_desktop(job_data):
    # Define desktop directory path
    desktop_path = os.path.join(os.path.expanduser("~"), "Desktop")

    # Define CSV file name
    csv_file = os.path.join(desktop_path, "output.csv")

    # Define column names
    columns = ["job_title", "category"]


    # Write to CSV file
    with open(csv_file, 'w', newline='') as file:
        writer = csv.writer(file)
        writer.writerow(columns)
        writer.writerows(job_data)

    print(f"CSV file has been saved to {csv_file}")
```

```python
import csv
import os
from datetime import datetime

def data_to_csv_full_desktop(job_data):
    # Define desktop directory path
    desktop_path = os.path.join(os.path.expanduser("~"), "Desktop")

    # Define timestamp
    timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")

    # Define CSV file name with timestamp
    csv_file = os.path.join(desktop_path, f"results_{timestamp}.csv")

    # Define column names
    columns = ["job_title", "company_name", "work_genre", "area", "income", "description", "url"]

    # Write to CSV file
    with open(csv_file, 'w', newline='') as file:
        writer = csv.writer(file)
        writer.writerow(columns)
        writer.writerows(job_data)

    print(f"CSV file has been saved to {csv_file}")
```
